```python
import os
import sys
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_settings_v3 import *
from auxiliary import get_base_network_name, read_data_file

sys.path.append('fast')
from functions import percolate, edgelist_to_nn_set

%matplotlib inline

networks_dir = os.path.join('..', 'networks')

def get_r_from_meank(meank, N):

    # Ignoring borders
    #r = np.sqrt((1/np.pi)*((meank-1)/(N-1)))
    r = np.sqrt((1/np.pi)*((meank)/(N)))
    return r
```

```python
expected_k = 6

prefactors = {
    1024: 1.0255,
    2048: 1.013,
    4096: 1.01,
    8192: 1.0065,
    16384: 1.0026
}

prefactors = {
    1024: 1.0203,
    2048: 1.0135,
    4096: 1.009,
    8192: 1.0068,
    16384: 1.0035
}

r_values = {N: prefactors[N]*get_r_from_meank(expected_k, N)  for N in prefactors.keys()}
r_values
```

```python
ncols = 2
fig, axes = plt.subplots(figsize=(8*ncols, 6), ncols=ncols)

ax = axes[0]
ax.set_xscale('log')
ax.set_yscale('log')

X = np.array(list(prefactors.keys()))
Y = np.array(list(prefactors.values()))
ax.plot(X, Y, 'o-')

ax = axes[1]
ax.set_xscale('log')
ax.set_yscale('log')

X = np.array(list(r_values.keys()))
Y = np.array(list(r_values.values()))
ax.plot(X, Y, 'o-')
```
